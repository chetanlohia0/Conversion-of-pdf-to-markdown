```python
import gc
import os
import re
import tempfile
import traceback
from datetime import datetime
from pathlib import Path
from collections import Counter, defaultdict
from math import ceil

import fitz
import torch
from transformers import AutoTokenizer, AutoModel

MODEL_NAME = "baidu/Unlimited-OCR"
MODEL = None
TOKENIZER = None

def gpu_status():
    if not torch.cuda.is_available():
        print("CUDA unavailable")
        return
    allocated = torch.cuda.memory_allocated() / 1024**2
    reserved = torch.cuda.memory_reserved() / 1024**2
    print(f"Allocated : {allocated:.2f} MB")
    print(f"Reserved  : {reserved:.2f} MB")

def clear_gpu():
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.synchronize()
        torch.cuda.empty_cache()
        torch.cuda.ipc_collect()
    gc.collect()

gpu_status()
```

```python
def load_model():
    global MODEL, TOKENIZER
    if MODEL is not None:
        print("Model already loaded.")
        return
    print("Loading model...")
    TOKENIZER = AutoTokenizer.from_pretrained(MODEL_NAME, trust_remote_code=True)
    MODEL = AutoModel.from_pretrained(
        MODEL_NAME,
        trust_remote_code=True,
        use_safetensors=True,
        torch_dtype=torch.bfloat16,
    )
    MODEL = MODEL.eval().cuda()
    print("Model loaded.")
    gpu_status()

def destroy_model():
    global MODEL, TOKENIZER
    print("Destroying model...")
    if MODEL is not None:
        del MODEL
    if TOKENIZER is not None:
        del TOKENIZER
    MODEL = None
    TOKENIZER = None
    clear_gpu()
    print("Model destroyed.")
    gpu_status()
```

```python
def pdf_to_images(pdf_path, dpi=300):
    doc = fitz.open(pdf_path)
    temp_dir = Path(tempfile.mkdtemp(prefix="pdf_pages_"))
    pages = []
    matrix = fitz.Matrix(dpi / 72, dpi / 72)
    for page_number, page in enumerate(doc):
        img_path = temp_dir / f"page_{page_number+1:04d}.png"
        page.get_pixmap(matrix=matrix).save(str(img_path))
        pages.append(str(img_path))
    doc.close()
    return pages
```

```python
def batch_output_folder(batch):
    folder = OUTPUT_ROOT / (
        f"batch_{batch['id']}_pages_{batch['start']:03d}_{batch['end']:03d}"
    )
    folder.mkdir(parents=True, exist_ok=True)
    return folder

def validate_output(out_dir, expected_pages):
    md_file = out_dir / "result.md"
    if not md_file.exists():
        return False, "Missing result.md"
    text = md_file.read_text(encoding="utf8", errors="ignore")
    if len(text.strip()) < 100:
        return False, "Output too small"
    page_count = text.count("<PAGE>")
    if page_count != expected_pages:
        return False, f"Expected {expected_pages} pages but found {page_count}"
    word_count = len(text.split())
    if word_count < expected_pages * 30:
        return False, f"Too few words ({word_count})"
    has_table = "<table>" in text
    has_non_text = "[Non-Text]" in text
    if not has_table and not has_non_text:
        print("Warning: No tables or non-text regions detected.")
    return True, "OK"

SUCCESS = "SUCCESS"
OOM = "OOM"
ERROR = "ERROR"
INVALID = "INVALID"
SKIPPED = "SKIPPED"

def log_batch(batch, status, message=""):
    out_dir = batch_output_folder(batch)
    log_file = out_dir / "batch.log"
    with open(log_file, "a", encoding="utf8") as f:
        f.write("=" * 60 + "\n")
        f.write(f"Time   : {datetime.now()}\n")
        f.write(f"Batch  : {batch['id']}\n")
        f.write(f"Pages  : {batch['start']} - {batch['end']}\n")
        f.write(f"Status : {status}\n")
        if message:
            f.write(f"Message: {message}\n")
        f.write("\n")
```

```python
def process_batch(batch):
    out_dir = batch_output_folder(batch)
    done_file = out_dir / ".done"
    if done_file.exists():
        print(f"✓ Batch {batch['id']} already completed.")
        return SKIPPED

    print("\n" + "="*70)
    print(f"Batch {batch['id']} ({batch['start']} -> {batch['end']})")
    print("="*70)

    try:
        MODEL.infer_multi(
            TOKENIZER,
            prompt="<image>Multi page parsing.",
            image_files=batch["pages"],
            output_path=str(out_dir),
            image_size=1024,
            max_length=32768,
            no_repeat_ngram_size=35,
            ngram_window=1024,
            save_results=True,
        )
    except torch.OutOfMemoryError:
        clear_gpu()
        log_batch(batch, OOM, "CUDA Out Of Memory")
        return OOM
    except RuntimeError as e:
        if "out of memory" in str(e).lower():
            clear_gpu()
            log_batch(batch, OOM, str(e))
            return OOM
        traceback.print_exc()
        clear_gpu()
        log_batch(batch, ERROR, str(e))
        return ERROR
    except Exception as e:
        traceback.print_exc()
        clear_gpu()
        log_batch(batch, ERROR, str(e))
        return ERROR

    clear_gpu()
    valid, reason = validate_output(out_dir, len(batch["pages"]))
    if not valid:
        print(reason)
        log_batch(batch, INVALID, reason)
        return INVALID

    log_batch(batch, SUCCESS)
    done_file.touch()
    print("✓ Batch completed.")
    return SUCCESS
```

```python
def split_batch(batch):
    pages = batch["pages"]
    if len(pages) == 1:
        return []
    mid = len(pages) // 2
    left_pages = pages[:mid]
    right_pages = pages[mid:]
    left = {
        "id": batch["id"] + "A",
        "start": batch["start"],
        "end": batch["start"] + len(left_pages) - 1,
        "pages": left_pages
    }
    right = {
        "id": batch["id"] + "B",
        "start": left["end"] + 1,
        "end": batch["end"],
        "pages": right_pages
    }
    return [left, right]

FAILED_BATCHES = []

def process_failed_batches():
    global FAILED_BATCHES
    while FAILED_BATCHES:
        batch = FAILED_BATCHES.pop(0)
        print("\n" + "="*70)
        print(f"Recovering Batch {batch['id']}")
        print(f"Pages : {batch['start']} -> {batch['end']}")
        print(f"Remaining Queue : {len(FAILED_BATCHES)}")
        print("="*70)
        children = split_batch(batch)
        if len(children) == 0:
            print("Single page still failed.")
            out = batch_output_folder(batch)
            (out / ".failed").touch()
            continue
        for child in children:
            status = process_batch(child)
            clear_gpu()
            if status in [SUCCESS, SKIPPED]:
                continue
            FAILED_BATCHES.append(child)
```

```python
def merge_markdown():
    batch_dirs = list(OUTPUT_ROOT.glob("batch_*"))
    print(f"\nFound {len(batch_dirs)} batch folders")
    batches = []
    for folder in batch_dirs:
        done_file = folder / ".done"
        md_file = folder / "result.md"
        if not done_file.exists() or not md_file.exists():
            continue
        m = re.search(r"batch_(B[0-9A-Z]+)_pages_(\d+)_(\d+)", folder.name)
        if m is None:
            continue
        batches.append({
            "id": m.group(1),
            "start": int(m.group(2)),
            "end": int(m.group(3)),
            "folder": folder
        })

    leaf_batches = []
    ids = [b["id"] for b in batches]
    for batch in batches:
        is_parent = any(other.startswith(batch["id"]) and other != batch["id"] for other in ids)
        if not is_parent:
            leaf_batches.append(batch)

    leaf_batches.sort(key=lambda x: x["start"])

    print("\nLeaf batches used for merge:")
    for b in leaf_batches:
        print(f"{b['id']} : {b['start']} -> {b['end']}")

    merged = []
    for batch in leaf_batches:
        md = batch["folder"] / "result.md"
        with open(md, "r", encoding="utf8", errors="ignore") as f:
            merged.append(f.read().strip())

    merged_file = OUTPUT_ROOT / "merged.md"
    with open(merged_file, "w", encoding="utf8") as f:
        f.write("\n\n".join(merged))

    print("\nMerged File Saved")
    print(merged_file)
    return merged_file

def document_statistics():
    merged = OUTPUT_ROOT / "merged.md"
    text = merged.read_text(encoding="utf8", errors="ignore")
    print("="*60)
    print("Document Statistics")
    print("="*60)
    print("Pages :", text.count("<PAGE>"))
    print("Tables :", text.count("<table>"))
    print("Non Text :", text.count("[Non-Text]"))
    print("Words :", len(text.split()))
    print("Characters :", len(text))
```

```python
# ============================================================
# Markdown Optimizer — Stage 1: Load, Clean, Split
# ============================================================
# NOTE: INPUT_MD now points directly at the merged.md produced
# by the OCR stage above (OUTPUT_ROOT). No manual re-upload needed.

INPUT_MD = OUTPUT_ROOT / "merged.md"
CLEANED_MD_PATH = OUTPUT_ROOT / "cleaned.md"

print("="*60)
print("Markdown Optimizer — Stage 1: Load, Clean, Split")
print("="*60)
print("Input :", INPUT_MD)

def load_markdown(path):
    if not path.exists():
        raise FileNotFoundError(path)
    text = path.read_text(encoding="utf8", errors="ignore")
    print(f"Characters : {len(text):,}")
    print(f"Words      : {len(text.split()):,}")
    return text

RAW_MD = load_markdown(INPUT_MD)
```

```python
# Matches a line that is ENTIRELY a single bracket-enclosed tag,
# e.g. "[Non-Text]", "[ Non Text ]", "[Blank]", "[Illegible]".
# Does NOT touch brackets embedded inside real sentences
# (e.g. "See section [3.2]"), since those are content, not
# OCR-engine metadata.
OCR_TAG_LINE = re.compile(
    r'^\s*\[\s*[A-Za-z][A-Za-z\-\s]{1,25}\s*\]\s*$'
)

def strip_ocr_artifact_lines(text):
    lines = text.splitlines()
    kept = []
    removed = []
    for line in lines:
        if OCR_TAG_LINE.match(line):
            removed.append(line.strip())
        else:
            kept.append(line)
    return "\n".join(kept), removed

CLEANED_MD, REMOVED_TAGS = strip_ocr_artifact_lines(RAW_MD)
tag_counts = Counter(REMOVED_TAGS)

print("\n" + "="*60)
print("OCR Artifact Tag Removal")
print("="*60)
print(f"Lines removed : {len(REMOVED_TAGS)}")
print(f"Unique tags   : {len(tag_counts)}")
for tag, count in tag_counts.most_common():
    print(f"  {count:>3} x  {tag}")

print(f"\nCharacters before : {len(RAW_MD):,}")
print(f"Characters after  : {len(CLEANED_MD):,}")

CLEANED_MD_PATH.write_text(CLEANED_MD, encoding="utf8")
print(f"\nSaved cleaned md : {CLEANED_MD_PATH}")
```

```python
PAGE_DELIM = re.compile(r'\s*<PAGE>\s*', re.IGNORECASE)

def split_pages(markdown):
    chunks = PAGE_DELIM.split(markdown)
    pages = [c.strip() for c in chunks if c.strip()]
    print(f"Pages : {len(pages)}")
    return pages

PAGE_TEXTS = split_pages(CLEANED_MD)   # uses CLEANED_MD, not RAW_MD
```

```python
PAGES = []
for i, page_text in enumerate(PAGE_TEXTS):
    raw_lines = page_text.splitlines()
    lines = [ln.rstrip() for ln in raw_lines if ln.strip()]
    PAGES.append({
        "page": i + 1,
        "text": page_text,
        "raw_lines": raw_lines,
        "lines": lines,
        "line_count": len(lines),
        "word_count": len(page_text.split()),
        "char_count": len(page_text),
    })

total_lines = sum(p["line_count"] for p in PAGES)
total_words = sum(p["word_count"] for p in PAGES)
total_chars = sum(p["char_count"] for p in PAGES)

print("\n" + "="*60)
print("Document Statistics (post-cleaning)")
print("="*60)
print(f"Pages            : {len(PAGES)}")
print(f"Lines            : {total_lines:,}")
print(f"Words            : {total_words:,}")
print(f"Characters       : {total_chars:,}")
print(f"Avg Lines/Page   : {total_lines/len(PAGES):.1f}")
print(f"Avg Words/Page   : {total_words/len(PAGES):.1f}")
print(f"Estimated Tokens : {int(total_chars/4):,}")
```

```python
from rapidfuzz import fuzz

# ============================================================
# CONFIG
# ============================================================
SEARCH_RATIO = 0.40
MIN_LINES = 5
MAX_LINES = 15
MAX_NGRAM = 3
FUZZY_THRESHOLD = 80
TOKEN_SIM_THRESHOLD = 70
MAX_DIFFERING_TOKENS = 2

# ============================================================
# STEP 1: BUILD SEARCH REGIONS
# ============================================================
for page in PAGES:
    lines = page["lines"]
    total = len(lines)
    region = max(MIN_LINES, min(MAX_LINES, ceil(total * SEARCH_RATIO)))
    page["region_size"] = region
    page["top_region"] = lines[:region]
    page["bottom_region"] = lines[-region:]

# ============================================================
# STEP 2: NORMALIZE (with digit masking)
# ============================================================
def normalize(text):
    text = text.lower()
    text = re.sub(r'\d+', '<NUM>', text)
    text = re.sub(r'[^a-z<>0-9_ ]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

for page in PAGES:
    page["top_norm"] = [normalize(l) for l in page["top_region"] if normalize(l)]
    page["bottom_norm"] = [normalize(l) for l in page["bottom_region"] if normalize(l)]
```

```python
RAW_CANDIDATES = defaultdict(lambda: {"count": 0, "pages": set(), "occurrences": []})

def collect_ngrams(region_lines, region_name, page_no):
    for n in range(1, MAX_NGRAM + 1):
        for i in range(len(region_lines) - n + 1):
            phrase = " ".join(region_lines[i:i+n]).strip()
            if not phrase:
                continue
            entry = RAW_CANDIDATES[phrase]
            entry["count"] += 1
            entry["pages"].add(page_no)
            entry["occurrences"].append({
                "page": page_no, "region": region_name, "start": i, "end": i + n - 1
            })

for page in PAGES:
    collect_ngrams(page["top_norm"], "top", page["page"])
    collect_ngrams(page["bottom_norm"], "bottom", page["page"])

print(f"Raw candidates (exact-match): {len(RAW_CANDIDATES)}")
```

```python
# ============================================================
# STEP 4: TOKEN-LEVEL MERGE GUARD
# ============================================================
def tokenize_for_diff(text):
    return re.split(r'[\s_]+', text.strip())

def token_diff_is_noise(text_a, text_b):
    tokens_a = tokenize_for_diff(text_a)
    tokens_b = tokenize_for_diff(text_b)
    if len(tokens_a) != len(tokens_b):
        return False
    diffs = [(ta, tb) for ta, tb in zip(tokens_a, tokens_b) if ta != tb]
    if len(diffs) == 0:
        return True
    if len(diffs) > MAX_DIFFERING_TOKENS:
        return False
    for ta, tb in diffs:
        if ta == "<num>" or tb == "<num>":
            continue
        sim = fuzz.ratio(ta, tb)
        min_len = min(len(ta), len(tb))
        local_threshold = TOKEN_SIM_THRESHOLD if min_len > 4 else 90
        if sim < local_threshold:
            return False
    return True

# ============================================================
# STEP 5: FUZZY CLUSTER CANDIDATES (with guard)
# ============================================================
def cluster_candidates(raw_candidates, whole_threshold=FUZZY_THRESHOLD):
    items = sorted(raw_candidates.items(), key=lambda kv: kv[1]["count"], reverse=True)
    clusters = []
    for text, info in items:
        if len(text) < 3:
            continue
        placed = False
        for cluster in clusters:
            whole_score = fuzz.ratio(text, cluster["canonical"])
            if whole_score >= whole_threshold and token_diff_is_noise(text, cluster["canonical"]):
                cluster["members"].append(text)
                cluster["count"] += info["count"]
                cluster["pages"].update(info["pages"])
                cluster["occurrences"].extend(info["occurrences"])
                placed = True
                break
        if not placed:
            clusters.append({
                "canonical": text,
                "members": [text],
                "count": info["count"],
                "pages": set(info["pages"]),
                "occurrences": list(info["occurrences"]),
            })
    return clusters

CLUSTERS = cluster_candidates(RAW_CANDIDATES)
print(f"Clusters after fuzzy merge (with token guard): {len(CLUSTERS)}")
```

```python
TOTAL_PAGES = len(PAGES)
for c in CLUSTERS:
    c["page_ratio"] = len(c["pages"]) / TOTAL_PAGES
    top_n = sum(1 for o in c["occurrences"] if o["region"] == "top")
    bot_n = sum(1 for o in c["occurrences"] if o["region"] == "bottom")
    c["region_guess"] = "header" if top_n >= bot_n else "footer"

CLUSTERS.sort(key=lambda c: (len(c["pages"]), len(c["canonical"])), reverse=True)

print("="*100)
print("TOP CLUSTERS BY PAGE COVERAGE")
print("="*100)
for c in CLUSTERS[:30]:
    print("-"*100)
    print(f"CANONICAL   : {c['canonical']}")
    print(f"VARIANTS    : {sorted(set(c['members']))[:5]}")
    print(f"PAGES       : {len(c['pages'])}/{TOTAL_PAGES}  (ratio={c['page_ratio']:.2f})")
    print(f"GUESS       : {c['region_guess']}")
```

```python
COVERAGE_THRESHOLD = 0.40
LINE_MATCH_THRESHOLD = 80   # lenient — used only against pre-validated signatures

ACCEPTED = [c for c in CLUSTERS if c["page_ratio"] >= COVERAGE_THRESHOLD]
HEADER_CLUSTERS = [c for c in ACCEPTED if c["region_guess"] == "header"]
FOOTER_CLUSTERS = [c for c in ACCEPTED if c["region_guess"] == "footer"]

def consolidate_signatures(clusters):
    texts = sorted({c["canonical"] for c in clusters}, key=len, reverse=True)
    final = []
    for t in texts:
        if any(t != kept and t in kept for kept in final):
            continue  # already covered by a longer signature
        final.append(t)
    return final

HEADER_SIGNATURES = consolidate_signatures(HEADER_CLUSTERS)
FOOTER_SIGNATURES = consolidate_signatures(FOOTER_CLUSTERS)

print("="*60)
print(f"Accepted clusters (ratio >= {COVERAGE_THRESHOLD}): {len(ACCEPTED)}")
print("="*60)
print(f"\nFinal HEADER signatures ({len(HEADER_SIGNATURES)}):")
print("-"*60)
for h in sorted(HEADER_SIGNATURES, key=len, reverse=True):
    print(f"  {h}")
print(f"\nFinal FOOTER signatures ({len(FOOTER_SIGNATURES)}):")
print("-"*60)
for f in sorted(FOOTER_SIGNATURES, key=len, reverse=True):
    print(f"  {f}")
```

```python
def line_matches_any(norm_line, signatures, threshold=LINE_MATCH_THRESHOLD):
    if not norm_line:
        return False
    for sig in signatures:
        score = fuzz.partial_ratio(norm_line, sig)
        if score >= threshold:
            return True
    return False

for page in PAGES:
    lines = page["lines"]
    region_size = page["region_size"]
    total = len(lines)
    top_idx_range = range(0, min(region_size, total))
    bottom_start = max(0, total - region_size)
    bottom_idx_range = range(bottom_start, total)

    drop_indices = set()
    matched_header_lines = []
    matched_footer_lines = []

    for i in top_idx_range:
        norm = normalize(lines[i])
        if line_matches_any(norm, HEADER_SIGNATURES):
            drop_indices.add(i)
            matched_header_lines.append(lines[i])

    for i in bottom_idx_range:
        norm = normalize(lines[i])
        if line_matches_any(norm, FOOTER_SIGNATURES):
            drop_indices.add(i)
            matched_footer_lines.append(lines[i])

    kept_lines = [ln for i, ln in enumerate(lines) if i not in drop_indices]
    page["cleaned_lines"] = kept_lines
    page["cleaned_text"] = "\n".join(kept_lines)
    page["removed_header_lines"] = matched_header_lines
    page["removed_footer_lines"] = matched_footer_lines
    page["removed_count"] = len(drop_indices)

total_removed = sum(p["removed_count"] for p in PAGES)
total_before = sum(p["line_count"] for p in PAGES)

print("\n" + "="*60)
print("Removal Complete (fuzzy line-level)")
print("="*60)
print(f"Lines before : {total_before}")
print(f"Lines removed: {total_removed}")
print(f"Lines after  : {total_before - total_removed}")

print("\nTop 10 pages by removal %:")
for p in sorted(PAGES, key=lambda p: p["removed_count"]/max(p["line_count"],1), reverse=True)[:10]:
    pct = p["removed_count"] / max(p["line_count"], 1) * 100
    print(f"  Page {p['page']:>3} | {p['line_count']:>3} lines | removed {p['removed_count']:>3} ({pct:.0f}%)")
```
